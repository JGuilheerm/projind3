***IMPORTAÇÃO***

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/JGuilheerm/projind3/main/dados.csv')
```

```python
df = pd.DataFrame(dados)
```

```python
dados
```

***DADOS, CALCULOS, GRAFICOS E ANALISES***

Total de horas trabalhadas:

```python
horasTrabalhadas = df['Horas Trabalhadas'].sum()
print(f'Horas trabalhadas: {horasTrabalhadas}')
```

Média diária de horas trabalhadas:

```python
mediaHorasTrabalhadas = df['Horas Trabalhadas'].mean()
print(f'Média diária de horas trabalhadas: {mediaHorasTrabalhadas:.2f}h')
```

Total de bugs corrigidos:

```python
bugsCorrigidos = df['Bugs Corrigidos'].sum()
print(f' Total de bugs corrigidos: {bugsCorrigidos}')
```

Médio diária de bugs corrigidos:

```python
print(f'Média de bugs corrigidos por dia: {bugsCorrigidos/7:.1f}')
```

Total de tarefas concluídas:

```python
tarefasConcluidas = df['Tarefas Concluídas'].sum()
print(f'Total de tarefas concluídas: {tarefasConcluidas}')
```

Média de tarefas concluída por dia:

```python
mediaTarefasConcluidas = df['Tarefas Concluídas'].mean()
print(f'Média diária de tarefas concluídas é de: {mediaTarefasConcluidas:.1f} ')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Tarefas Concluídas'], marker='o')
plt.title('Tarefas Concluidas')
plt.xlabel('Dias da Semana')
plt.ylabel('Tarefas Concluidas')
plt.grid(True)
plt.show()
```

OBS: Uma boa consistência no numero de tarefas concluídas sugera uma boa gestão na equipe.

Produtividade diária:

```python
produtividade = tarefasConcluidas / horasTrabalhadas
print(f'Produtividade diária de: {produtividade:.1f} tarefas concluídas por hora')
```

```python
df['Produtividade'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Produtividade'], marker='o', color='y')
plt.title('Produtividade Diária')
plt.xlabel('Dias da Semana')
plt.ylabel('Tarefas Concluídas por Hora')
plt.grid(True)
plt.show()
```

 OBS: Medindo a eficiencia do trabalho ao longo da semana, podemos ver uma media consideravel na produtividade, embora seja necessario investigar as razões da queda brusca na produtividade do dia 1.
